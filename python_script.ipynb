```python
#Import our Libraries
import pandas as pd
import sqlite3
```

```python
#Read in our CSV
df = pd.read_csv('customer_orders.csv')
```

```python
df.columns
```

```python
df_unique_columns = df[['order number', 'employee id', 'employee name',
       'employee_job_title', 'order date', 'order type', 'customer type',
       'cust id', 'customer name', 'prod category',
       'prod number', 'prod_name', 'quantity', 'price', 'order total']]
```

```python
df_unique_columns
```

```python
df_clean_columns = df_unique_columns.reset_index()
df_clean_columns 
```

```python
df_clean_columns.columns = df_clean_columns.columns.str.replace(" ", "")
df_clean_columns.head(10)
```

```python
df_clean_columns[df.duplicated()]
```

```python
df_clean_columns_rows = df_clean_columns.drop_duplicates(subset=["ordernumber", "orderdate"])
df_clean_columns_rows.head()
```

```python
connection = sqlite3.connect("database.db")
```

```python
df_clean_columns_rows.to_sql("customer_orders_table", connection, if_exists="replace", index=False)
```
